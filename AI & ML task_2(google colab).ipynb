```python
#corerlation matrix to feature relationship
import seaborn as sns
import matplotlib.pyplot as plt

# Create correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Titanic Features")
plt.show()

#pairplot for seleacted features
# Use a smaller subset for readability
selected = df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']]

# Pairplot
sns.pairplot(selected, hue='Survived', palette='Set2')
plt.suptitle("Pairplot of Titanic Features (by Survival)", y=1.02)
plt.show()
```
